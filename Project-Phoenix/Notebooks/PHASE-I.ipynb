# PHASE I: Control Notebook

```python
import time
from run_objects.scan import scan
from run_objects.sys3CS import sys3CS
import pyfirmata
import os
```

```python
# Scan Params
Duration       = 30  #seconds
Raman_duration = 0.5   #seconds
Iterations     = 12

# Paths
path_to_phase = "D:\\CC Hargy\\Project-Phoenix\\PHASE-I"
```

```python
# Functions
def take_exposure(sys, time, savepath=None):
    
    # Set exposure time
    sys.ctrl(spec_exp=time)
    
    # Open laser shutter
    boardMAR.digital[13].write(1)
    
    # Open spectro shutter
    sys.ctrl(spec_shutter=True)
    
    # Run
    sys.ctrl(spec_run=True)
    
    # Close both shutters
    sys.ctrl(spec_shutter=False)
    boardMAR.digital[13].write(0)
    
    if savepath != None:
            sys.ctrl(spec_save=savepath)
    
    sys.ctrl(spec_clear_screen=True)

def take_ambience(sys, time, savepath=None):
    
    # Set exposure time
    sys.ctrl(spec_exp=time)
    
    # Make sure laser shutter closed
    boardMAR.digital[13].write(0)
    
    # Open spectro shutter
    sys.ctrl(spec_shutter=True)
    
    # Run
    sys.ctrl(spec_run=True)
    
    # Close shutter
    sys.ctrl(spec_shutter=False)
    
    if savepath != None:
            sys.ctrl(spec_save=savepath)
    
    sys.ctrl(spec_clear_screen=True)

def take_baseline(sys, time, savepath=None):
    
    # Set exposure time
    sys.ctrl(spec_exp=time)
    
    # Make sure laser shutter closed
    boardMAR.digital[13].write(0)
    
    # Make sure spectro shutter closed
    sys.ctrl(spec_shutter=False)
    
    # Run
    sys.ctrl(spec_run=True)
    
    if savepath != None:
            sys.ctrl(spec_save=savepath)
    
    sys.ctrl(spec_clear_screen=True)
```

```python
# instantiate object
sys = sys3CS()
boardMAR = pyfirmata.Arduino("COM23")
sys.ctrl(spec_wl=690)
```

```python
boardMAR.digital[13].write(0)
```
